```python
import pandas as pd
import numpy as np

# df is your feedback dataframe
# faculty_domain_avg is computed earlier
# domains list already exists

df_feedback = df.copy()   # rename for clarity
```

```python
# Calculate overall SD of ratings across all 15 domains
all_ratings = df_feedback[domains].values.flatten()

std_dev = np.std(all_ratings)
RSA = 1 - (std_dev / 5)

print(" Rating Stability Accuracy (RSA):", round(RSA * 100, 2), "%")
```

```python
def cronbach_alpha(df):
    df_scores = df[domains]
    item_vars = df_scores.var(axis=0, ddof=1)
    total_var = df_scores.sum(axis=1).var(ddof=1)
    k = df_scores.shape[1]
    alpha = (k / (k - 1)) * (1 - item_vars.sum() / total_var)
    return alpha

alpha = cronbach_alpha(df_feedback)
print(" Domain Reliability (Cronbach Alpha):", round(alpha * 100, 2), "%")
```

```python
faculty_scores = faculty_domain_avg.mean(axis=1)
inst_avg = faculty_scores.mean()

deviation = abs(faculty_scores - inst_avg).mean()
FEA = 1 - (deviation / 5)

print(" Faculty Evaluation Accuracy (FEA):", round(FEA * 100, 2), "%")
```

```python
!pip install textblob
from textblob import TextBlob
```

```python
def sentiment(text):
    if pd.isna(text) or text.strip() == "":
        return 0
    return TextBlob(text).sentiment.polarity

df_feedback["sentiment"] = df_feedback["feedback_comment"].apply(sentiment)
```

```python
aligned = 0

for _, row in df_feedback.iterrows():
    rat = np.mean([row[d] for d in domains])  # avg domain score
    sent = row["sentiment"]

    if sent > 0 and rat >= 3:     # positive comment  high rating
        aligned += 1
    elif sent < 0 and rat < 3:   # negative comment  low rating
        aligned += 1
    elif -0.1 <= sent <= 0.1 and 2.5 <= rat <= 3.5:  # neutral  mid rating
        aligned += 1

SRA = aligned / len(df_feedback)

print(" SentimentRating Alignment Accuracy (SRA):", round(SRA * 100, 2), "%")
```

```python
overall_accuracy = (RSA + alpha + FEA + SRA) / 4

print("\n OVERALL SYSTEM ACCURACY:", round(overall_accuracy * 100, 2), "%")
```

```python
def cronbach_alpha(df):
    df_scores = df[domains]
    item_vars = df_scores.var(axis=0, ddof=1)
    total_var = df_scores.sum(axis=1).var(ddof=1)
    k = df_scores.shape[1]
    alpha = (k / (k - 1)) * (1 - item_vars.sum() / total_var)
    return alpha

alpha = cronbach_alpha(df_feedback)
print(" Domain Reliability (Cronbach Alpha):", round(alpha * 100, 2), "%")
```

## TECHNICAL ALGORITHM FOR FEEDBACK ANALYSIS

 **ALGORITHM NAME:**
***Faculty Feedback Multi-Domain Weighted Evaluation Algorithm (FF-MDWEA)***

 **INPUT:**

F = Set of faculty, size = N

S = Set of students, size = M

D = 15-domain rating vector per feedback

R = Rating given by student i to faculty j

R(d) = Rating of domain d for faculty j

 **OUTPUT:**

Faculty Performance Vector  FP(j)

Domain Strength Vector  DS(d)

Best/Weakest Faculty

Strongest/Weakest Domain

Visualization matrices (Heatmap, Radar, Summary Indicators)

 **ALGORITHM STEPS (Technical + Mathematical)**

**STEP 1  Data Normalization**

Purpose: Remove bias between rating patterns.

For each domain d:

$R^*_{i,j,d} = \frac{R_{i,j,d} - \min(R_d)}{\max(R_d) - \min(R_d)}$

This converts ratings  normalized range [0,1].

If normalization unnecessary, skip step.

**STEP 2  Domain Aggregation (Faculty-Level)**

For each faculty j, compute domain-wise mean:

$D_{avg}(j,d) = \frac{1}{M} \sum_{i=1}^{M} R_{i,j,d}$

Where

M = number of students who rated that faculty (100)

d = 1 to 15 domains

**STEP 3  Weighted Faculty Score**

Each domain can have weight W(d) (optional).
(Default = equal  1)

$FP(j) = \frac{\sum_{d=1}^{15} (D_{avg}(j,d) \times W(d))}{\sum_{d=1}^{15} W(d)}$

FP(j) = overall performance score of faculty j.

**STEP 4  Domain Strength (Institute-Level)**

Calculate average domain performance across all faculties:

$DS(d) = \frac{1}{N} \sum_{j=1}^{N} D_{avg}(j,d)$

Where

N = number of faculty (20)

Higher DS(d)  stronger teaching domain
Lower DS(d)  weakest domain  training required

**STEP 5  Ranking Algorithm**
Best Faculty:
$BestFaculty = \arg\max_{j \in F} FP(j)$
